```python
import csv
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
county_name = "Yuba"
number_of_elections = 3
number_of_seats = 5
```

Stats csv is generated in maup_chain_county run

```python
stats = pd.read_csv('Yuba_stats.csv')
```

```python
stats
```

```python
stats1 = stats[['2016_pres_d_wins','2016_ussn_harris_wins','2016_cg_d_wins']]
```

```python
fig, ax = plt.subplots()

plt.title("Estimated Democratic Seat Share for " + county_name + " County")
plt.xlabel('Seats')
plt.ylabel('Frequency')

for i in range(number_of_elections):

    # defining some local variables
    election_name = stats1.keys()[i]
    width = 1.0/(number_of_elections+1)
    offset = width * (float(number_of_elections-1)/2 - i )

    # preprocessing step: turning the list of 0s, 1s, ... 5s into
    #  a list (of length 6) saying how many 0s, 1s, ... there are.
    heights_i, bins_i = np.histogram( stats1[election_name], 
                            bins= number_of_seats+1, 
                            range= (0,number_of_seats+1) )

    # building the bar chart for this election
    #  and adding it to the legend.
    election = ax.bar(bins_i[:-1] + offset, heights_i, width=width) #facecolor='green'
    election.set_label("using " + election_name[0:4] + " " + election_name[5:9] + " data")
    ax.legend()

plt.show()
```
